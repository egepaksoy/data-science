```python
import pandas as pd
sales = pd.read_csv("https://raw.githubusercontent.com/egepaksoy/datasets/master/sales_data.csv")
```

```python
youth_persons_cost = sales.loc[(sales['Age_Group'] == 'Youth (<25)') & (sales['Cost'] >= 255)]['Cost'].shape[0], "Youth"
senior_persons_cost = sales.loc[("Seniors (64+)" == sales['Age_Group']) & (sales['Cost'] >= 255)]['Cost'].shape[0], "Seniors"
adult_persons_cost = sales.loc[("Adults (35-64)" == sales['Age_Group']) & (sales['Cost'] >= 255)]['Cost'].shape[0], "Adults"
yadult_persons_cost = sales.loc[("Young Adults (25-34)" == sales['Age_Group']) & (sales['Cost'] >= 255)]['Cost'].shape[0], "Y. Adults"
```

```python
best = youth_persons_cost
age_costs = [youth_persons_cost, senior_persons_cost, adult_persons_cost, yadult_persons_cost]

for age in age_costs:
  if age[0] > best[0]:
    best = age

print(best)
```
